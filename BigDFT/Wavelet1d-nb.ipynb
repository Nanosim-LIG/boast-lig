```ruby
require './GenericConvolution.rb'
require './WaveletFilters.rb'
```

```ruby
def Wavelet1d(wavelet_filter, direction, optims=GenericOptimization::new)
  conv_operation = GenericConvolutionOperator1d::new(wavelet_filter, :wavelet => direction, :a => true, :a_y => true, :ld => true, :narr => true)
  conv_operation.optimize(optims) if optims

  p, subops = conv_operation.procedure

  kernel = BOAST::CKernel::new

  print_header

  subops.each_value { |op|
    BOAST::pr op
    puts "chosen:"+ op.name
  }
  BOAST::pr p

  kernel.procedure = p
  kernel.cost_function = lambda { |*args| conv_operation.cost(*args) }
  return kernel
end
```

```ruby
optims = GenericOptimization::new(:unroll_range => [6,6], :mod_arr_test => false, :tt_arr_test => false)
wave_filter = WaveletFilter::new("sym#{SYM8_LP.length/2}", SYM8_LP)
5
```

```ruby
ksym8 = Wavelet1d( wave_filter, :decompose, optims )
5
```

```ruby
ksym8i = Wavelet1d( wave_filter, :recompose, optims )
```

```ruby
n1 = 62
n2 = 66
n3 = 65
ndims=NArray.int(3)
ndims[0]=n1
ndims[1]=n2
ndims[2]=n3
type = NArray::FLOAT
epsilon = 10e-12
input   = ANArray::new(type, 32, n1*2, n2*2, n3*2).random!
work1  = ANArray::new(type, 32, n1*2, n2*2, n3*2)
output  = ANArray::new(type, 32, n1*2, n2*2, n3*2)
output2 = ANArray::new(type, 32, n1*2, n2*2, n3*2)

scal=NArray.float(3)
hgrid = NArray.float(3)
hgrid[0] = 0.5
hgrid[1] = 0.6
hgrid[2] = 0.7
(0..2).each{ |ind| scal[ind] = -0.5 / (hgrid[ind]*hgrid[ind])   }
5
```

```ruby
ksym8.run(3, 0, ndims, BC::PERIODIC, ndims, ndims, 1, input, output, scal[0],0.0)
ksym8.run(3, 1, ndims, BC::PERIODIC, ndims, ndims, 1, output, work1, scal[1],0.0)
ksym8.run(3, 2, ndims, BC::PERIODIC, ndims, ndims, 1, work1, output, scal[2],0.0)
```

```ruby
ksym8i.run(3, 0, ndims, BC::PERIODIC, ndims, ndims, 1, output, output2, 1.0/scal[0],0.0)
ksym8i.run(3, 1, ndims, BC::PERIODIC, ndims, ndims, 1, output2, work1, 1.0/scal[1],0.0)
ksym8i.run(3, 2, ndims, BC::PERIODIC, ndims, ndims, 1, work1, output2, 1.0/scal[2],0.0)
```

```ruby
diff = (input - output2).abs
diff.each { |elem|
  raise "Warning: residue too big: #{elem}" if elem > epsilon
}
puts 'ok'
```
